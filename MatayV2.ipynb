```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import RobustScaler, LabelEncoder
from sklearn.model_selection import train_test_split
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM, Dropout, Bidirectional
from tensorflow.keras.layers import LayerNormalization, Conv1D
from sklearn.metrics import classification_report, confusion_matrix, precision_recall_curve, roc_curve, auc
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')

print(f"TensorFlow version: {tf.__version__}")

# Veriyi okuma
df = pd.read_csv('/content/drive/MyDrive/MATAY_DATA/transposed_timestamp_value_value1_machine_id_9_with_faultsV2_train.csv', on_bad_lines='skip')


# Zaman damgasını düzenleme
df['timestamp'] = pd.to_datetime(df['timestamp'], format='ISO8601', errors='coerce')

# Veri seti analizi
print("Fault value counts:")
print(df['Fault'].value_counts())

fault_indices = df[df['Fault'] == -99].index
print("\nFault timestamps:")
for idx in fault_indices:
    print(f"Fault at index {idx}, timestamp: {df['timestamp'].iloc[idx]}")

# Etiketleri dönüştür (-99 -> 1, 0 -> 0)
df['Fault'] = (df['Fault'] == -99).astype(int)

# Kullanılacak özellikleri seçme
features = ['Current L1', 'Current L2', 'Current L3',
           'Voltage L1-L2', 'Voltage L2-L3', 'Voltage L3-L1',
           'ΣActive Power', 'ΣApparent Power', 'ΣPower Factor',
           'Current Unbalance ', 'Measured Frequency', 'Neutral Current',
           'Total Consumed Energy', 'Total Consumed Reactive Energy',
           'Voltage Unbalance']

def add_engineered_features(df, features):
    """Türetilmiş özellikler ekleme"""
    df_new = df.copy()

    for feature in features:
        # Hareketli ortalamalar ve standart sapmalar
        df_new[f'{feature}_ma_5'] = df_new[feature].rolling(window=5).mean()
        df_new[f'{feature}_std_5'] = df_new[feature].rolling(window=5).std()

        # Değişim oranları
        df_new[f'{feature}_delta'] = df_new[feature].diff()

        # Hareketli varyans
        df_new[f'{feature}_var_5'] = df_new[feature].rolling(window=5).var()

        # Hareketli medyan
        df_new[f'{feature}_median_5'] = df_new[feature].rolling(window=5).median()

    return df_new

# Veri ön işleme ve özellik mühendisliği
df_engineered = add_engineered_features(df.copy(), features)
df_engineered = df_engineered.fillna(method='ffill').fillna(method='bfill')

# X ve y verilerini hazırlama
X = df_engineered.drop(['timestamp', 'Fault'], axis=1).values
y = df_engineered['Fault'].values

# Veriyi normalize etme
scaler = RobustScaler()
X_scaled = scaler.fit_transform(X)

def create_fault_focused_sequences(data, labels, seq_length, pre_fault_window=100):
    """Hata odaklı sekanslar oluştur - geliştirilmiş versiyon"""
    X, y = [], []
    fault_indices = np.where(labels == 1)[0]  # -99 yerine 1 kullanıyoruz

    print(f"Found {len(fault_indices)} fault instances")

    # Her hata için birden fazla sekans oluştur
    for idx in fault_indices:
        if idx < pre_fault_window + seq_length:
            continue

        # Hata öncesi farklı noktalardan örnekler al
        for start_offset in range(pre_fault_window):
            if start_offset % 5 == 0:  # Her 5 noktada bir örnek al
                start_idx = idx - pre_fault_window + start_offset
                if start_idx + seq_length <= idx:
                    sequence = data[start_idx:start_idx + seq_length]
                    if not np.any(np.isnan(sequence)):
                        X.append(sequence)
                        y.append(1)  # -99 yerine 1 kullanıyoruz

    print(f"Created {len(X)} fault sequences")

    # Normal örnekler için geliştirilmiş örnekleme
    normal_indices = []
    current_idx = 0

    for fault_idx in fault_indices:
        safe_end = max(0, fault_idx - pre_fault_window)
        if current_idx < safe_end:
            # Normal durumlar için daha yoğun örnekleme
            step = max(1, (safe_end - current_idx) // 1000)
            normal_indices.extend(range(current_idx, safe_end, step))
        current_idx = fault_idx + seq_length

    # Son bölgeden örnekler
    if current_idx < len(data) - seq_length:
        step = max(1, (len(data) - seq_length - current_idx) // 1000)
        normal_indices.extend(range(current_idx, len(data) - seq_length, step))

    # Normal örnekleri dengele
    n_fault_sequences = len(X)
    n_normal_sequences = min(len(normal_indices), n_fault_sequences * 3)  # 3:1 oranı
    selected_normal_indices = np.random.choice(normal_indices, n_normal_sequences, replace=False)

    for idx in selected_normal_indices:
        sequence = data[idx:idx + seq_length]
        if not np.any(np.isnan(sequence)):
            X.append(sequence)
            y.append(0)

    print(f"Added {n_normal_sequences} normal sequences")

    X = np.array(X)
    y = np.array(y)

    # Veriyi karıştır
    indices = np.arange(len(X))
    #np.random.shuffle(indices)

    return X[indices], y[indices]

def custom_loss(y_true, y_pred):
    """Geliştirilmiş özel kayıp fonksiyonu"""
    # Binary cross entropy
    bce = tf.keras.losses.binary_crossentropy(y_true, y_pred)

    # Hatalara daha fazla ağırlık ver (artık 1 kullanıyoruz)
    weights = tf.where(y_true == 1, 30.0, 1.0)

    # Hata yakınındaki tahminlere kademeli ağırlık ver
    confidence_penalty = tf.where(
        tf.abs(y_true - y_pred) > 0.7,
        3.0,  # Büyük hatalar için daha yüksek ceza
        tf.where(
            tf.abs(y_true - y_pred) > 0.3,
            2.0,  # Orta seviye hatalar için orta ceza
            1.0   # Küçük hatalar için normal ağırlık
        )
    )

    return bce * weights * confidence_penalty

# Model parametreleri
SEQUENCE_LENGTH = 20
BATCH_SIZE = 32
EPOCHS = 150
VALIDATION_SPLIT = 0.2
TEST_SPLIT = 0.2
PRE_FAULT_WINDOW = 150

# Veri setini hazırla
X_seq, y_seq = create_fault_focused_sequences(X_scaled, y, SEQUENCE_LENGTH, PRE_FAULT_WINDOW)

# Veri seti dağılımını kontrol et
print("\nSequence label distribution:")
unique, counts = np.unique(y_seq, return_counts=True)
print(dict(zip(unique, counts)))

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(
    X_seq, y_seq,
    test_size=TEST_SPLIT,
    stratify=y_seq
    #random_state=42
)

def create_improved_model(sequence_length, n_features):
    """Geliştirilmiş model mimarisi"""
    model = Sequential([
        # 1D Konvolüsyon katmanları
        Conv1D(filters=32, kernel_size=3, padding='same', activation='relu',
               input_shape=(sequence_length, n_features)),
        LayerNormalization(),
        Dropout(0.2),

        Conv1D(filters=64, kernel_size=3, padding='same', activation='relu'),
        LayerNormalization(),
        Dropout(0.2),

        # Bidirectional LSTM katmanları
        Bidirectional(LSTM(64, return_sequences=True)),
        LayerNormalization(),
        Dropout(0.3),

        Bidirectional(LSTM(32)),
        LayerNormalization(),
        Dropout(0.3),

        # Dense katmanları
        Dense(64, activation='relu'),
        LayerNormalization(),
        Dropout(0.2),

        Dense(32, activation='relu'),
        LayerNormalization(),
        Dropout(0.2),

        Dense(1, activation='sigmoid')
    ])

    optimizer = tf.keras.optimizers.Adam(
        learning_rate=0.001,
        beta_1=0.9,
        beta_2=0.999,
        epsilon=1e-07
    )

    model.compile(
        optimizer=optimizer,
        loss=custom_loss,
        metrics=['accuracy', tf.keras.metrics.AUC()]
    )

    return model

# Callbacks
callbacks = [
    tf.keras.callbacks.EarlyStopping(
        monitor='val_loss',
        patience=20,
        restore_best_weights=True,
        min_delta=0.001
    ),
    tf.keras.callbacks.ReduceLROnPlateau(
        monitor='val_loss',
        factor=0.5,
        patience=10,
        min_lr=0.00001,
        verbose=1
    )
]

# Model oluşturma
model = create_improved_model(SEQUENCE_LENGTH, X_scaled.shape[1])
model.summary()

# Class weights hesaplama
n_normal = np.sum(y_train == 0)
n_fault = np.sum(y_train == 1)  # Artık 1 kullanıyoruz
class_weights = {
    0: 1.,
    1: (n_normal / n_fault) * 2  # Hata sınıfına daha fazla ağırlık
}

# Model eğitimi
history = model.fit(
    X_train, y_train,
    epochs=EPOCHS,
    batch_size=BATCH_SIZE,
    validation_split=VALIDATION_SPLIT,
    callbacks=callbacks,
    class_weight=class_weights,
    verbose=1
)

def evaluate_predictions(y_true, y_pred, threshold=0.5):
    """Geliştirilmiş değerlendirme metrikleri"""
    y_pred_binary = (y_pred > threshold).astype(int)

    print("Classification Report:")
    print(classification_report(y_true, y_pred_binary))

    print("\nConfusion Matrix:")
    print(confusion_matrix(y_true, y_pred_binary))

    precision, recall, pr_thresholds = precision_recall_curve(y_true, y_pred)
    pr_auc = auc(recall, precision)

    fpr, tpr, roc_thresholds = roc_curve(y_true, y_pred)
    roc_auc = auc(fpr, tpr)

    # Görselleştirme
    plt.figure(figsize=(15, 5))

    # Precision-Recall eğrisi
    plt.subplot(1, 2, 1)
    plt.plot(recall, precision, label=f'PR curve (AUC = {pr_auc:.2f})')
    plt.xlabel('Recall')
    plt.ylabel('Precision')
    plt.title('Precision-Recall Curve')
    plt.grid(True)
    plt.legend()

    # ROC eğrisi
    plt.subplot(1, 2, 2)
    plt.plot(fpr, tpr, label=f'ROC curve (AUC = {roc_auc:.2f})')
    plt.plot([0, 1], [0, 1], 'k--')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('ROC Curve')
    plt.grid(True)
    plt.legend()

    plt.tight_layout()
    plt.show()

    return {
        'pr_auc': pr_auc,
        'roc_auc': roc_auc,
        'precision': precision,
        'recall': recall,
        'fpr': fpr,
        'tpr': tpr
    }

# Model değerlendirme
predictions = model.predict(X_test)
metrics = evaluate_predictions(y_test, predictions)

# Eğitim geçmişini görselleştirme
plt.figure(figsize=(15, 5))

plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Model Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True)

plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Model Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()

# Modeli kaydet
model.save('improved_welding_fault_detection_model.h5')
```

```python
import pandas as pd
import numpy as np
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.preprocessing import RobustScaler
import tensorflow as tf
from tensorflow.keras.models import load_model
import matplotlib.pyplot as plt
import seaborn as sns
from tabulate import tabulate

def custom_loss(y_true, y_pred):
    """Model yüklemek için gerekli custom loss fonksiyonu"""
    bce = tf.keras.losses.binary_crossentropy(y_true, y_pred)
    weights = tf.where(y_true == 1, 30.0, 1.0)
    confidence_penalty = tf.where(
        tf.abs(y_true - y_pred) > 0.7,
        3.0,
        tf.where(
            tf.abs(y_true - y_pred) > 0.3,
            2.0,
            1.0
        )
    )
    return bce * weights * confidence_penalty

def prepare_features(df, features):
    """Özellikleri hazırla"""
    df_features = df.copy()

    # Her özellik için istatistikler hesapla
    for feature in features:
        # Hareketli ortalamalar
        df_features[f'{feature}_ma_5'] = df_features[feature].rolling(window=5, min_periods=1).mean()
        df_features[f'{feature}_std_5'] = df_features[feature].rolling(window=5, min_periods=1).std()

        # Değişim oranları
        df_features[f'{feature}_delta'] = df_features[feature].diff().fillna(0)

        # Hareketli varyans ve medyan
        df_features[f'{feature}_var_5'] = df_features[feature].rolling(window=5, min_periods=1).var()
        df_features[f'{feature}_median_5'] = df_features[feature].rolling(window=5, min_periods=1).median()

    return df_features

def create_sequences(data, sequence_length):
    """Veriyi sequence'lere dönüştür"""
    sequences = []
    for i in range(len(data) - sequence_length + 1):
        sequence = data[i:(i + sequence_length)]
        sequences.append(sequence)
    return np.array(sequences)

def evaluate_multi_step_predictions(model, X_test, y_test, steps=[1, 5, 10], thresholds=[0.3, 0.5, 0.7]):
    """Farklı adımlar için tahmin performansını değerlendir"""
    results = []

    sequence_length = X_test.shape[1]  # sequence uzunluğu

    for step in steps:
        print(f"\nEvaluating {step}-step ahead predictions...")

        # Test verisini kaydır
        y_shifted = y_test[step:]  # step kadar kaydırılmış gerçek değerler
        X_current = X_test[:-step]  # Mevcut input verisi

        if len(X_current) == 0 or len(y_shifted) == 0:
            print(f"Not enough data for {step}-step prediction")
            continue

        # Tahminleri al
        predictions = model.predict(X_current, verbose=0)

        # Her threshold için metrikleri hesapla
        for threshold in thresholds:
            y_pred = (predictions > threshold).astype(int)

            # Boyutları eşitle
            min_len = min(len(y_pred), len(y_shifted))
            y_pred = y_pred[:min_len]
            y_true = y_shifted[:min_len]

            # Metrikleri hesapla
            metrics = {
                'Steps': step,
                'Threshold': threshold,
                'Accuracy': accuracy_score(y_true, y_pred),
                'Precision': precision_score(y_true, y_pred, zero_division=0),
                'Recall': recall_score(y_true, y_pred, zero_division=0),
                'F1': f1_score(y_true, y_pred, zero_division=0)
            }
            results.append(metrics)

            print(f"\nMetrics for {step}-step ahead, threshold {threshold}:")
            print(f"Accuracy: {metrics['Accuracy']:.4f}")
            print(f"Precision: {metrics['Precision']:.4f}")
            print(f"Recall: {metrics['Recall']:.4f}")
            print(f"F1 Score: {metrics['F1']:.4f}")

    # Sonuçları DataFrame'e dönüştür
    df_results = pd.DataFrame(results)

    # Sonuçları görselleştir
    plot_results(df_results)

    return df_results

def plot_results(df_results):
    """Sonuçları görselleştir"""
    metrics = ['Accuracy', 'Precision', 'Recall', 'F1']
    plt.figure(figsize=(20, 5))

    for i, metric in enumerate(metrics, 1):
        plt.subplot(1, 4, i)

        for threshold in df_results['Threshold'].unique():
            mask = df_results['Threshold'] == threshold
            plt.plot(df_results[mask]['Steps'],
                    df_results[mask][metric],
                    marker='o',
                    label=f'Threshold={threshold}')

        plt.title(f'{metric} vs Prediction Steps')
        plt.xlabel('Steps Ahead')
        plt.ylabel(metric)
        plt.grid(True)
        plt.legend()

    plt.tight_layout()
    plt.show()

def main():
    # Model yükle
    print("Loading model...")
    model = load_model('improved_welding_fault_detection_model.h5',
                      custom_objects={'custom_loss': custom_loss})

    # Test verisini oku
    print("Loading test data...")
    test_df = pd.read_csv('/content/drive/MyDrive/MATAY_DATA/transposed_timestamp_value_value1_machine_id_9_with_faultsV2_test.csv')

    # Özellikleri belirle
    features = ['Current L1', 'Current L2', 'Current L3',
                'Voltage L1-L2', 'Voltage L2-L3', 'Voltage L3-L1',
                'ΣActive Power', 'ΣApparent Power', 'ΣPower Factor',
                'Current Unbalance ', 'Measured Frequency', 'Neutral Current',
                'Total Consumed Energy', 'Total Consumed Reactive Energy',
                'Voltage Unbalance']

    # Özellikleri hazırla
    print("Preparing features...")
    df_features = prepare_features(test_df, features)

    # Eksik değerleri doldur
    df_features = df_features.fillna(method='ffill').fillna(method='bfill')

    # Etiketleri hazırla (-99 -> 1, 0 -> 0)
    y = (test_df['Fault'] == -99).astype(int)

    # Feature kolonlarını seç
    feature_cols = df_features.columns.difference(['timestamp', 'Fault'])

    # Normalizasyon
    print("Normalizing features...")
    scaler = RobustScaler()
    X_scaled = scaler.fit_transform(df_features[feature_cols])

    # Sequence'leri oluştur
    print("Creating sequences...")
    sequence_length = 20  # model input shape'inden alınmalı
    X_test = create_sequences(X_scaled, sequence_length)
    y_test = y[sequence_length-1:]

    print(f"Input shape: {X_test.shape}")
    print(f"Output shape: {y_test.shape}")

    # Farklı tahmin ufukları için değerlendir
    steps = [1, 5, 10]  # tahmin adımları
    thresholds = [0.3, 0.5, 0.7]  # threshold değerleri

    # Sonuçları hesapla
    results = evaluate_multi_step_predictions(model, X_test, y_test, steps, thresholds)

    # Sonuçları tablo olarak göster
    print("\nDetailed Results:")
    print(tabulate(results, headers='keys', tablefmt='pretty', floatfmt='.4f'))

    # En iyi sonuçları göster
    print("\nBest Results for Each Step:")
    best_results = results.loc[results.groupby('Steps')['F1'].idxmax()]
    print(tabulate(best_results, headers='keys', tablefmt='pretty', floatfmt='.4f'))

    # Sonuçları kaydet
    results.to_csv('multi_step_prediction_results.csv', index=False)
    best_results.to_csv('best_prediction_results.csv', index=False)

if __name__ == "__main__":
    main()
```

```python
import pandas as pd
import numpy as np
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.preprocessing import RobustScaler
import tensorflow as tf
from tensorflow.keras.models import load_model
import matplotlib.pyplot as plt
import seaborn as sns
from tabulate import tabulate

def custom_loss(y_true, y_pred):
    """Model yüklemek için gerekli custom loss fonksiyonu"""
    bce = tf.keras.losses.binary_crossentropy(y_true, y_pred)
    weights = tf.where(y_true == 1, 30.0, 1.0)
    confidence_penalty = tf.where(
        tf.abs(y_true - y_pred) > 0.7,
        3.0,
        tf.where(
            tf.abs(y_true - y_pred) > 0.3,
            2.0,
            1.0
        )
    )
    return bce * weights * confidence_penalty

def prepare_features(df, features):
    """Özellikleri hazırla"""
    df_features = df.copy()

    # Her özellik için istatistikler hesapla
    for feature in features:
        # Hareketli ortalamalar
        df_features[f'{feature}_ma_5'] = df_features[feature].rolling(window=5, min_periods=1).mean()
        df_features[f'{feature}_std_5'] = df_features[feature].rolling(window=5, min_periods=1).std()

        # Değişim oranları
        df_features[f'{feature}_delta'] = df_features[feature].diff().fillna(0)

        # Hareketli varyans ve medyan
        df_features[f'{feature}_var_5'] = df_features[feature].rolling(window=5, min_periods=1).var()
        df_features[f'{feature}_median_5'] = df_features[feature].rolling(window=5, min_periods=1).median()

    return df_features

def create_sequences(data, sequence_length):
    """Veriyi sequence'lere dönüştür"""
    sequences = []
    for i in range(len(data) - sequence_length + 1):
        sequence = data[i:(i + sequence_length)]
        sequences.append(sequence)
    return np.array(sequences)

def evaluate_multi_step_predictions(model, X_test, y_test, steps=[1, 5, 10], thresholds=[0.3, 0.5, 0.7]):
    """Farklı adımlar için tahmin performansını değerlendir"""
    results = []

    sequence_length = X_test.shape[1]  # sequence uzunluğu

    for step in steps:
        print(f"\nEvaluating {step}-step ahead predictions...")

        # Test verisini kaydır
        y_shifted = y_test[step:]  # step kadar kaydırılmış gerçek değerler
        X_current = X_test[:-step]  # Mevcut input verisi

        if len(X_current) == 0 or len(y_shifted) == 0:
            print(f"Not enough data for {step}-step prediction")
            continue

        # Tahminleri al
        predictions = model.predict(X_current, verbose=0)

        # Her threshold için metrikleri hesapla
        for threshold in thresholds:
            y_pred = (predictions > threshold).astype(int)

            # Boyutları eşitle
            min_len = min(len(y_pred), len(y_shifted))
            y_pred = y_pred[:min_len]
            y_true = y_shifted[:min_len]

            # Metrikleri hesapla
            metrics = {
                'Steps': step,
                'Threshold': threshold,
                'Accuracy': accuracy_score(y_true, y_pred),
                'Precision': precision_score(y_true, y_pred, zero_division=0),
                'Recall': recall_score(y_true, y_pred, zero_division=0),
                'F1': f1_score(y_true, y_pred, zero_division=0)
            }
            results.append(metrics)

            print(f"\nMetrics for {step}-step ahead, threshold {threshold}:")
            print(f"Accuracy: {metrics['Accuracy']:.4f}")
            print(f"Precision: {metrics['Precision']:.4f}")
            print(f"Recall: {metrics['Recall']:.4f}")
            print(f"F1 Score: {metrics['F1']:.4f}")

    # Sonuçları DataFrame'e dönüştür
    df_results = pd.DataFrame(results)

    # Sonuçları görselleştir
    plot_results(df_results)

    return df_results

def plot_results(df_results):
    """Sonuçları görselleştir"""
    metrics = ['Accuracy', 'Precision', 'Recall', 'F1']
    plt.figure(figsize=(20, 5))

    for i, metric in enumerate(metrics, 1):
        plt.subplot(1, 4, i)

        for threshold in df_results['Threshold'].unique():
            mask = df_results['Threshold'] == threshold
            plt.plot(df_results[mask]['Steps'],
                    df_results[mask][metric],
                    marker='o',
                    label=f'Threshold={threshold}')

        plt.title(f'{metric} vs Prediction Steps')
        plt.xlabel('Steps Ahead')
        plt.ylabel(metric)
        plt.grid(True)
        plt.legend()

    plt.tight_layout()
    plt.show()

def main():
    # Model yükle
    print("Loading model...")
    model = load_model('improved_welding_fault_detection_model.h5',
                      custom_objects={'custom_loss': custom_loss})

    # Test verisini oku
    print("Loading test data...")
    test_df = pd.read_csv('/content/drive/MyDrive/MATAY_DATA/transposed_timestamp_value_value1_machine_id_9_with_faultsV2_test.csv')

    # Özellikleri belirle
    features = ['Current L1', 'Current L2', 'Current L3',
                'Voltage L1-L2', 'Voltage L2-L3', 'Voltage L3-L1',
                'ΣActive Power', 'ΣApparent Power', 'ΣPower Factor',
                'Current Unbalance ', 'Measured Frequency', 'Neutral Current',
                'Total Consumed Energy', 'Total Consumed Reactive Energy',
                'Voltage Unbalance']

    # Özellikleri hazırla
    print("Preparing features...")
    df_features = prepare_features(test_df, features)

    # Eksik değerleri doldur
    df_features = df_features.fillna(method='ffill').fillna(method='bfill')

    # Etiketleri hazırla (-99 -> 1, 0 -> 0)
    y = (test_df['Fault'] == -99).astype(int)

    # Feature kolonlarını seç
    feature_cols = df_features.columns.difference(['timestamp', 'Fault'])

    # Normalizasyon
    print("Normalizing features...")
    scaler = RobustScaler()
    X_scaled = scaler.fit_transform(df_features[feature_cols])

    # Sequence'leri oluştur
    print("Creating sequences...")
    sequence_length = 20  # model input shape'inden alınmalı
    X_test = create_sequences(X_scaled, sequence_length)
    y_test = y[sequence_length-1:]

    print(f"Input shape: {X_test.shape}")
    print(f"Output shape: {y_test.shape}")

    # Farklı tahmin ufukları için değerlendir
    steps = [5, 10, 20]  # tahmin adımları
    thresholds = [0.5, 0.7, 0.9]  # threshold değerleri

    # Sonuçları hesapla
    results = evaluate_multi_step_predictions(model, X_test, y_test, steps, thresholds)

    # Sonuçları tablo olarak göster
    print("\nDetailed Results:")
    print(tabulate(results, headers='keys', tablefmt='pretty', floatfmt='.4f'))

    # En iyi sonuçları göster
    print("\nBest Results for Each Step:")
    best_results = results.loc[results.groupby('Steps')['F1'].idxmax()]
    print(tabulate(best_results, headers='keys', tablefmt='pretty', floatfmt='.4f'))

    # Sonuçları kaydet
    results.to_csv('multi_step_prediction_results.csv', index=False)
    best_results.to_csv('best_prediction_results.csv', index=False)

if __name__ == "__main__":
    main()
```
